Pick up a dataset of your choice for clustering ([UCI dataset](https://archive.ics.uci.edu/ml/datasets.php)). Apply various clustering algorithms (including k-means and spectral clustering) to this dataset and explain your observation by using insightful plots or other some other metrics.

```python
# Your code comes here
Iris = pd.read_csv("Iris.csv")
```

```python
Iris["Class"].replace({"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2}, inplace = True)
iris = Iris.values
```

```python
#K-Means
model = KMeans(n_clusters=2)
# fit the model
model.fit(iris)
# assign a cluster to each example
yhat = model.predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 0], iris[row_ix, 1], s=500)
# show the plot
plt.show()

```

```python
#Spectral Clustering
model = SpectralClustering(n_clusters=2)
# fit model and predict clusters
yhat = model.fit_predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 1], iris[row_ix, 2], s = 500)
# show the plot
plt.show()
```

```python
#Mean Shift
model = MeanShift()
# fit model and predict clusters
yhat = model.fit_predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 2], iris[row_ix, 3], s=500)
# show the plot
plt.show()

```

```python
#Affinity Propogation
model = AffinityPropagation(damping=0.9)
# fit the model
model.fit(iris)
# assign a cluster to each example
yhat = model.predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 1], iris[row_ix, 2], s=500)
# show the plot
plt.show()
```

```python
#BIRCH
model = Birch(threshold=0.01, n_clusters=2)
# fit the model
model.fit(iris)
# assign a cluster to each example
yhat = model.predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 1], iris[row_ix, 3], s=500)
# show the plot
plt.show()
```

```python
#OPTICS
model = OPTICS(eps=0.8, min_samples=10)
# fit the model
model.fit(iris)
# assign a cluster to each example
yhat = model.fit_predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 0], iris[row_ix, 2], s=500)
# show the plot
plt.show()
```

```python
#Gaussian Mixture
model = GaussianMixture(n_components=2)
# fit the model
model.fit(iris)
# assign a cluster to each example
yhat = model.fit_predict(iris)
# retrieve unique clusters
clusters = np.unique(yhat)
# create scatter plot for samples from each cluster
for cluster in clusters:
 # get row indexes for samples with this cluster
 row_ix = np.where(yhat == cluster)
 # create scatter of these samples
 plt.scatter(iris[row_ix, 0], iris[row_ix, 3], s=500)
# show the plot
plt.show()
```
