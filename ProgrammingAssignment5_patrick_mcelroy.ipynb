```python
# Import required packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

np.random.seed(100)

def display_image(arr_image,color_map=None,dim=(10,20)):
    arr_image = arr_image/np.max(arr_image)
    arr_image = np.uint8(arr_image*255)
    plt.figure(figsize=dim)
    plt.axis('off')
    plt.imshow(arr_image, cmap=color_map)
    plt.show()
```

Use the Python code for image compression to compress the national flags of different countries. The smalles rank $r$ gives the numerical rank of the flag. Pick up any 5 national flags and arrange them in the order of decreasing numerical rank.

```python
# Your code comes here
def pixCompress(pix, k):
    pix_rgb = pix/255.0;
    U0,S0,V0 = np.linalg.svd(pix_rgb[...,0]);
    U1,S1,V1 = np.linalg.svd(pix_rgb[...,1]);
    U2,S2,V2 = np.linalg.svd(pix_rgb[...,2]);

    pix_compressed = np.zeros_like(pix_rgb);
    pix_compressed[...,0] = np.dot(U0[:,:k], V0[:k,:]*S0[:k].reshape(k,1));
    pix_compressed[...,1] = np.dot(U1[:,:k], V1[:k,:]*S1[:k].reshape(k,1));
    pix_compressed[...,2] = np.dot(U2[:,:k], V2[:k,:]*S2[:k].reshape(k,1));
    
    print(f'Image compressed at k={k}:', pix_compressed.shape)
          
    display_image(pix_compressed,dim=(20,30))
```

```python
#First Flag Compressed
Leb = mpimg.imread("Lebanon.jpeg")
pixCompress(Leb, 230)
```

```python
#Second Flag Compressed
Mex = mpimg.imread("Mexico.jpeg")
pixCompress(Mex, 150)
```

```python
#Third Flag Compressed
US = mpimg.imread("US.jpeg")
pixCompress(US, 20)
```

```python
#Fourth Flag Compressed
Den = mpimg.imread("Danish.jpeg")
pixCompress(Den, 2)
```

```python
#Fifth Flag Compressed
France = mpimg.imread("France.jpeg")
pixCompress(France, 1)
```
